### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.dates as mdates

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# total_rides_city = pyber_data_df.groupby(["type"]).count()
# total_rides_city.loc[:,["ride_id"]]
total_rides_city = pyber_data_df.groupby(["type"]).count()['ride_id']
total_rides_city
```

```python
# 2. Get the total drivers for each city type
total_drivers_city = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_city
```

```python
#  3. Get the total amount of fares for each city type
total_fares_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city
```

```python
#  4. Get the average fare per ride for each city type. 
# avg_fares_city_df = pyber_data_df.groupby(["type"]).mean()['fare']
avg_fares_city = total_fares_city / total_rides_city
avg_fares_city
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fares_driver = total_fares_city / total_drivers_city
# pyber_data_df.groupby(["type"])
avg_fares_driver
# avg_fares_driver.loc[:,["fare"]]
```

```python
# avg_fares_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df_1 = pd.merge(total_rides_city, total_drivers_city, how="left", on=["type"])
pyber_summary_df_2 = pd.merge(pyber_summary_df_1, total_fares_city, how="left", on=["type"])

#insert column name for avg amounts then merge with other results
pyber_summary_df_3 = pd.merge(pyber_summary_df_2, avg_fares_city.rename('Average Fare per Ride'), how="left", on=["type"])
pyber_summary_df = pd.merge(pyber_summary_df_3, avg_fares_driver.rename('Average Fare per Driver'), how="left", on=["type"])

#rename the column name
pyber_summary_df.rename(columns = {'ride_id':'Total Rides'}, inplace = True)
pyber_summary_df.rename(columns = {'driver_count':'Total Drivers'}, inplace = True)
pyber_summary_df.rename(columns = {'fare':'Total Fares'}, inplace = True)

pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

#make a copy of original dataset
pyber_summary_df_format = pyber_summary_df.copy()

#convert all numbers with thousand comma and fares with 2 decimals, insert $ for fares
pyber_summary_df_format["Total Rides"] = pyber_summary_df_format["Total Rides"].map('{:,}'.format)
pyber_summary_df_format["Total Drivers"] = pyber_summary_df_format["Total Drivers"].map('{:,}'.format)
pyber_summary_df_format["Total Fares"] = "$" + pyber_summary_df_format["Total Fares"].map('{:,.2f}'.format)
pyber_summary_df_format["Average Fare per Ride"] = "$" + pyber_summary_df_format["Average Fare per Ride"].map('{:,.2f}'.format)
pyber_summary_df_format["Average Fare per Driver"] = "$" + pyber_summary_df_format["Average Fare per Driver"].map('{:,.2f}'.format)

pyber_summary_df_format
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_pyber_data = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
new_pyber_data_df = pd.DataFrame (new_pyber_data)
new_pyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
reset_pyber_data_df = new_pyber_data_df.reset_index()
reset_pyber_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_pyber_data_df = reset_pyber_data_df.pivot(values='fare', index=['date'], columns=['type'])
pivot_pyber_data_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

pivot_pyber_data_filter_df = pivot_pyber_data_df.loc[ \
                    (pd.to_datetime(pivot_pyber_data_df.index) > pd.to_datetime("2019-01-01")) & \
                    (pd.to_datetime(pivot_pyber_data_df.index) < pd.to_datetime("2019-04-29"))]
pivot_pyber_data_filter_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pivot_pyber_data_filter_df.index = pd.to_datetime(pivot_pyber_data_filter_df.index)
pivot_pyber_data_filter_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_pyber_data_filter_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pivot_pyber_data_filter_resample_df = pivot_pyber_data_filter_df.resample('w').sum()
pivot_pyber_data_filter_resample_df.head(10)
```

```python
final_summary = pivot_pyber_data_filter_resample_df.copy()
final_summary.index = pd.to_datetime(final_summary.index).month_name().str[:3]

# pd.to_datetime(pivot_pyber_data_df.index)
final_summary
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(16,5))


ax.plot(pivot_pyber_data_filter_resample_df.index, \
        pivot_pyber_data_filter_resample_df.Rural, \
        label="Rural", linewidth=2.5)
ax.plot(pivot_pyber_data_filter_resample_df.index, \
        pivot_pyber_data_filter_resample_df.Suburban, \
        label="Suburban", linewidth=2.5)
ax.plot(pivot_pyber_data_filter_resample_df.index, \
        pivot_pyber_data_filter_resample_df.Urban, \
        label="Urban", linewidth=2.5)

#format the graph
ax.set_yticks(np.arange(0, 2500, step=500))
ax.set_title("Total Fare by City Type", fontsize=20)
ax.set_ylabel("Fare ($USD)")

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="center", title="Type")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

#format xaxis 

# # Create minor ticks at an increment of 1.

ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1,2,3,4)))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.xaxis.set_minor_locator(mdates.MonthLocator())


# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
```
